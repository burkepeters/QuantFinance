# Central Limit

The normalized sum of independent random variables generally looks like a bell curve as sample size increases. 

We need a finite mean and stdev. 

In other words, the distribution of independent sample means is approximately normal (even if the population isn't). 

```python
import numpy as np
import random
import matplotlib.pyplot as plt
from scipy import stats
```

```python
population = np.random.normal(2, 2, 100000)
```

```python
sample_means = []
sample_num = [1000, 2000, 4000, 8000, 16000, 32000, 64000, 128000]
samplesize = 500
```

```python
for x in sample_num:
    sample_mean = []
    for j in range(0, x):
        rc = random.choices(population, k=samplesize)
        sample_mean.append(sum(rc) / len(rc))
    sample_means.append(sample_mean)
```

```python
cols = 2
rows = 4
fig, ax = plt.subplots(rows, cols, figsize=(20, 15))
n = 0
for i in range(0, rows):
    for j in range(0, cols):
        ax[i, j].hist(sample_means[n], 200, density=True)
        ax[i, j].set_title(label="number of sampling :" + str(sample_num[n]))
        n += 1
```

```python
np.mean(population)
```

```python
np.median(population)
```

```python
stats.mode(population)[0][0]
```

```python
population.std()
```

```python
stats.kurtosis(population)
```

```python
stats.skew(population)
```
